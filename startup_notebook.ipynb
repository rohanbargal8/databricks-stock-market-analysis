```python
file_path = "/Volumes/workspace/default/abc/stock_market_sample.csv"

df = spark.read.csv(file_path, header=True, inferSchema=True)

df.show()
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("stock_market_data")
```

```python
spark.sql("SELECT * FROM stock_market_data LIMIT 10").show()
```

```python
spark.sql("SELECT DISTINCT symbol FROM stock_market_data").show()
```

```python
spark.sql("""
SELECT trade_date, symbol, (close_price - open_price) AS price_change
FROM stock_market_data
ORDER BY trade_date
""").show()
```

```python
spark.sql("""
SELECT symbol, trade_date, volume
FROM stock_market_data
QUALIFY volume = MAX(volume) OVER (PARTITION BY symbol)
""").show()
```
